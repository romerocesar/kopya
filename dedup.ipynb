```python
from imagededup.methods import PHash, CNN
from imagededup.utils import plot_duplicates
```

```python
path = 'caltech-101/101_ObjectCategories/Faces/'
hasher = PHash()
```

```python
encodings = hasher.encode_images(image_dir=path)
```

```python
duplicates = hasher.find_duplicates(encoding_map=encodings)
```

```python
for original, dups in duplicates.items():
    if len(dups) < 3: continue
    plot_duplicates(image_dir=path, duplicate_map=duplicates, filename=original)
```

```python
cnn = CNN()
cnn_encodings = cnn.encode_images(image_dir=path)
cnn_duplicates = cnn.find_duplicates(encoding_map=cnn_encodings)
```

```python
for original, dups in cnn_duplicates.items():
    if len(dups) < 3: continue
    plot_duplicates(image_dir=path, duplicate_map=cnn_duplicates, filename=original)
```

```python
cnn.find_duplicates_to_remove(image_dir=path, encoding_map=cnn_encodings)
```
